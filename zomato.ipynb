```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts().values
```

```python
#pie chart
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.1f%%")
```

```python
df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### observation
# 1.when rating id btw 4.5. to 4.9 ----excellent
# 2.when rating id btw 4.0. to 4.4 ----v good
# 3.when rating id btw 3.5. to 3.9 ----good
# 4..
# 5..

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(17,10)  #increase the o/p image size
clr=list(ratings["Rating color"].unique())
clr[-1]='green'
clr[0]='blue'
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=clr)
```

### observations
# 1.not rated are very high
# 2.maximum no of ratings btw 2.5 to 3.4

```python
#count plot- for ploting categorical variables(frequency)
sns.countplot(x='Rating color',data=ratings,palette=clr)
```
